```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/uk-500.csv')
data
```

# Selection

## iloc

### Single Selection

```python
data.iloc[0]
```

```python
data.iloc[:,3]
```

### Selection of ranges across the dataframe using iloc

```python
# One row

# Range of rows
data.iloc[0:5]

# One column

# Range of columns

# One row, range of columns
data.iloc[:, 0:2]
# All rows, range of columns

# Specific set of rows and specific set of columns
data.iloc[[0,3,6,24], [0,5,6]]
data.iloc[0:5, 5:8]
```

## loc

### Label based / Index based indexing

```python
data.set_index('last_name', inplace=True)
data.head()
```

### Condition checks

```python
data.loc[['Andrade', 'Veness'], ['city','postal', 'phone1', 'email']]
```

```python
data.loc[((data['county']=='Kent') & (data['first_name']=='Aleshia')), 'city':'email']
```

```python
data.loc[data['company_name'].apply(lambda x: len(x.split(' ')) == 4)]
```
